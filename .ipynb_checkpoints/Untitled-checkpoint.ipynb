```python
import os
with open('new.m3u', 'w') as re:
    
    dicc = {}
    
    with open('live.m3u', 'r') as f:
        allcontext = f.read().split('\n')
        for i in range(len(allcontext) - 1):
            if allcontext[i][:1] == '#' and allcontext[i+1][:1] == 'h':
                kk = ''.join(allcontext[i].split(',')[1].split(' ')[0].split('-'))
                dicc[kk] = allcontext[i+1]
        
    # print(dicc)
        
    with open('demo.m3u', 'r') as f:
        l = f.readlines()
        # 输出M3U文件的每一行
        for i in range(len(l)):
            s = l[i].strip()
            if i % 2 == 1:
                re.write(s)
                re.write('\n')
                k = s.split('"')[1]
                print(k)
                if k in dicc:
                    re.write(dicc[k])
                    re.write('\n')
            else:
                if s != '':
                    re.write(s)
                    re.write('\n')
```
